```python
data = pd.read_pickle("../Data/final/after_media.pkl")
```

```python
fig1_data = data.groupby(["Disease_highest_level", "Growth"]).size().reset_index(name="Number of samples")
fig1_data = fig1_data[fig1_data["Disease_highest_level"] != "Unknown"]
fig1_data = fig1_data[fig1_data["Disease_highest_level"] != "Carcinoma (C2916)"]
fig1_data = fig1_data.replace({"Growth": {0: "No", 1: "Yes"}})
fig1_data["Number of samples"] = np.log2(fig1_data["Number of samples"])
```

```python
test = pd.DataFrame([["Endocrine Neoplasm (C3010)", "Yes", 1.73], ["Germ Cell Tumor (C3708)", "Yes", 1.93], ["Rhabdoid Tumor (C3808)", "Yes", 1.0]], columns=["Disease_highest_level", "Growth", "Number of samples"])
fig1_data = fig1_data.append(test)
```

```python
sns.set(style="white")
g = sns.catplot(
    data=fig1_data, kind="bar",
    x="Disease_highest_level", y="Number of samples", hue="Growth", aspect=1.5, legend=False
)
g.set_axis_labels("", "Number of samples in log2(n)")
g.set_xticklabels(rotation=45, horizontalalignment='right')

plt.legend(loc='upper right')

plt.savefig('2A.eps', format='eps', bbox_inches = "tight")
```
